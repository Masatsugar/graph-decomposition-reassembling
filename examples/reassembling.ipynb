```python
import rdkit.Chem as Chem
from rdkit.Chem import Draw
from rdkit import RDLogger

from mi_collections.chemutils import (
    get_mol,
    get_smiles,
    mol_with_atom_index,
    mol_from_graph,
    mol_to_graph,
)
from mi_collections.moldr.reassemble import merge_node, merge_edge, merge_mol_from_node, merge_mol_from_edge
```

```python
mol1 = mol_with_atom_index(get_mol("c1c2ccccc2ccc1"))
mol2 = mol_with_atom_index(get_mol("CC(=O)C1=NC=C(C)C=C1"))
mol = mol_with_atom_index(Chem.CombineMols(mol1, mol2))
Draw.MolToImage(mol)
```

```python
mol = merge_mol_from_node(mol1, mol2, 0, 10)
Draw.MolToImage(mol)
```

```python
mol = merge_mol_from_edge(mol1, mol2, (0, 9), (10, 15))
Draw.MolToImage(mol)
```

```python
molgraphs = merge_node(mol1, mol2)
_mols = [m.mol for m in molgraphs]
Draw.MolsToGridImage(_mols, molsPerRow=6, subImgSize=(300, 300))
```

```python
molgraphs = merge_edge(mol1, mol2)
_mols = [m.mol for m in molgraphs]
Draw.MolsToGridImage(_mols, molsPerRow=6, subImgSize=(300, 300))
```
